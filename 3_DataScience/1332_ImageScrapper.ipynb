```python
import requests
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen
import logging
import os
```

```python
save_dir = "images/"

if not os.path.exists(save_dir) :
    os.makedirs(save_dir)
```

```python
query = "beach"
response = requests.get(f"https://www.google.com/search?q={query}&sxsrf=AJOqlzV5v0uX79ane3ehlFgmedZ-Az-kNQ:1678289009982&source=lnms&tbm=isch&sa=X&ved=2ahUKEwjTz5Ti0cz9AhWRHXcKHZ3EDPEQ_AUoAXoECAEQAw&biw=1536&bih=714&dpr=1.25")
```

```python
response
```

```python
soup = bs(response.content, 'html.parser')
```

```python
images_tags = soup.find_all("img")
```

```python
del images_tags[0]
```

```python
len(images_tags)
```

```python
image_data_mongo = []

for i in images_tags : 
    image_url = i['src']
    image_data = requests.get(image_url).content
    mydict = {"index" : image_url, "image" : image_data}
    image_data_mongo.append(mydict)
    with open(os.path.join(save_dir, f"{query}_{images_tags.index(i)}.jpg"), "wb") as f:
        f.write(image_data)
        
```
