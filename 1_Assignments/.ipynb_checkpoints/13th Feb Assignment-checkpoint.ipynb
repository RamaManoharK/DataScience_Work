### Q1. Explain why we have to use the Exception class while creating a Custom Exception.

    In Python, exceptions are objects of the exception classes. All exception classes are the subclasses of the BaseException class.

    However, almost all built-in exception classes inherit from the Exception class, which is the subclass of the BaseException class
    
    custom exceptions - tailored to our specific use cases and that we can raise and catch in specific circumstances, it handles user defined specific errors only, will overlook other errors if any. Exception class will handle them.
    
    Exceptions must be directly or indirectly descended from the Exception class.

### Q2. Write a python program to print Python Exception Hierarchy.

```python
import inspect

def exception_hierarchy(cls, indent = 0) :
    print("." * indent, cls.__name__)
    for subcls in cls.__subclasses__() : # iterating through subclasses
        exception_hierarchy(subcls, indent + 3) # recursion

print("The class hierarchy for built-in exceptions is : ")

# building a tree hierarchy 
inspect.getclasstree(inspect.getmro(Exception))

exception_hierarchy(Exception)
```

### Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.

    The ArithmeticError Exception is the base class for all errors associated with arithmetic operations.
        ZeroDivisionError
        OverflowError
        FloatingPointError
    

```python
# ZeroDivisionError
# Raised when the second argument of a division or modulo operation is zero. The associated value is a string indicating the type of the operands and the operation

try :   
    10/0
except ZeroDivisionError as e :
    print("Error : {}".format(e))
```

```python
# OverFlowError :
# this occurs when an obtained value or result is larger than the declared operation or data type in the program
# then this throws an overflow error indicating the value is exceeding the given or declared limit value

import math

try :
    print("the exponential value is")
    print(math.exp(10000))
except OverflowError as e :
    print("After Overflow : {}".format(e))
```

### Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

    LookupError: when a key or index used on a mapping or sequence is invalid

```python
# IndexError: Raised when a sequence subscript is out of range. 
# Slice indices are silently truncated to fall in the allowed range; if an index is not an integer, TypeError is raised.

try : 
    ls = [2,3,4,5]
    print(ls[10])
except IndexError as e :
    print("Error : {}".format(e))
```

```python
# KeyError : Raised when a mapping (dictionary) key is not found in the set of existing keys.

try : 
    dict1 = {'name': 'rama', 'mail' : 'rama@gmail.com'}
    print(dict1['mobile'])

except KeyError as e : 
    print('key not found : {}'.format(e))
```

### Q5. Explain ImportError. What is ModuleNotFoundError?
    
    Raised when the import statement has troubles trying to load a module. Also raised when the “from list” in from ... import has a name that cannot be found.
    The name and path attributes can be set using keyword-only arguments to the constructor. When set they represent the name of the module that was attempted to be imported and the path to any file which triggered the exception, respectively.

    ModuleNotFoundError : 
        A subclass of ImportError which is raised by import when a module could not be located. It is also raised when None is found in sys.modules.
